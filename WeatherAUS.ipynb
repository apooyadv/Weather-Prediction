```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/home/apooyadv/Desktop/Mlib/weatherAUS.csv')
pd.set_option('display.max_columns', 23)
df.dropna(subset=['RainTomorrow'],inplace=True)
df
```

```python
plt.title('Count Vs Year')
plt.xlabel('Year')
plt.ylabel('Count')
sns.countplot(x = pd.to_datetime(df['Date']).dt.year)

pd.to_datetime(df['Date']).dt.year.value_counts()
```

```python
year = pd.to_datetime(df['Date']).dt.year

train_df = df[year < 2014]
val_df1 = df[year == 2014]
val_df2 = df[year == 2015]
test_df = df[year > 2015]

val_df = pd.concat([val_df1,val_df2])
```

```python
target_col = ['RainTomorrow']

train_targets = train_df[target_col]
val_targets = val_df[target_col]
test_targets  = test_df[target_col]
```

```python
input_cols = df.columns.tolist()[1:22]
```

```python
train_df = train_df[input_cols]
val_df = val_df[input_cols]
test_df = test_df[input_cols]
```

```python
numeric_cols = list(df[input_cols].select_dtypes(include=np.number))
cat_cols = list(df[input_cols].select_dtypes(include='object'))
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer()
imputer.fit(df[numeric_cols])
```

```python
train_df[numeric_cols] = imputer.transform(train_df[numeric_cols])
val_df[numeric_cols] = imputer.transform(val_df[numeric_cols])
test_df[numeric_cols] = imputer.transform(test_df[numeric_cols])
```

```python
train_df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(df[numeric_cols])
```

```python
train_df[numeric_cols] = scaler.transform(train_df[numeric_cols])
val_df[numeric_cols] = scaler.transform(val_df[numeric_cols])
test_df[numeric_cols] = scaler.transform(test_df[numeric_cols])
```

```python
train_df[cat_cols]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse=False,handle_unknown='ignore').fit(df[cat_cols])
```

```python
enc_cols = encoder.get_feature_names_out(cat_cols).tolist()
```

```python
train_df[enc_cols] = encoder.transform(train_df[cat_cols])
val_df[enc_cols] = encoder.transform(val_df[cat_cols])
test_df[enc_cols] = encoder.transform(test_df[cat_cols])
```

```python
train_df[enc_cols].drop(['WindDir3pm_nan','RainToday_nan','WindGustDir_nan','WindDir9am_nan'],axis=1,inplace=True)
val_df[enc_cols].drop(['WindDir3pm_nan','RainToday_nan','WindGustDir_nan','WindDir9am_nan'],axis=1,inplace=True)
test_df[enc_cols].drop(['WindDir3pm_nan','RainToday_nan','WindGustDir_nan','WindDir9am_nan'],axis=1,inplace=True)
```

```python
train_df = train_df[enc_cols + numeric_cols]
val_df = val_df[enc_cols + numeric_cols]
test_df = test_df[enc_cols + numeric_cols]
```

```python
def max_depth_cal(x):

    md = DecisionTreeClassifier(max_depth=x,random_state=80).fit(train_df,train_targets)
    train_err = 1-md.score(train_df,train_targets)
    val_err = 1-md.score(val_df,val_targets)

    return {'Max Depth':x,'Train Error':train_err,'Val Error':val_err}
```

```python
md_df = pd.DataFrame(max_depth_cal(x) for x in range(1,21))
```

```python
plt.plot(md_df['Max Depth'],md_df['Train Error'],label='Train Error')
plt.plot(md_df['Max Depth'],md_df['Val Error'],label='Val Error')
plt.title('Train Vs Val Error')
plt.xlabel('Max Depth')
plt.ylabel('Error')
plt.xticks(range(0,21,2))
plt.legend()
```

```python
def para_func(**kwargs):
    model = RandomForestClassifier(n_jobs=-1,random_state=80,**kwargs)
    model.fit(train_df,train_targets)

    return model.score(train_df,train_targets), model.score(val_df,val_targets)    
```
